```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
ax = plt.axes(projection='3d')

# 3次元折れ線グラフ用データ
zline = np.linspace(0, 20, 100)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')
print(zline)
print(xline)

# 3次元散布図用データ
zdata = 20 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

```python
def f(x, y):
  return -2*np.cos(8*np.pi*x/1)*np.cos(2*np.pi*(y+0.4)/0.8) -2*np.cos(6*np.pi*x/1)*np.cos(4*np.pi*(y+0.4)/0.8)

x = np.linspace(0, 1, 30)
y = np.linspace(0, 0.8, 30)

X, Y = np.meshgrid(x, y)
Z= f(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')

ax.view_init(45, -45)
```
